### 1. Смоделировать задачу о совпадающих днях рождения.

```python
def GetBirthday ():
    return randint(0,365)
def FirstMatch (size):
    result = [0]*size
    for i in range(size):
        a = [0]*366
        t = 1
        s = 0
        while a[t] < 2:
            t = GetBirthday()
            a[t] = a[t] + 1
            s = s + 1
        result[i] = s
    return result
```
