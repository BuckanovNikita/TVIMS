# Генерация выборки с заданной функцией распределения

Выбранное распределение f(x) = a * cos (x*a) на отрезке [0;pi/(2*a))

M[X] = (pi-2)/(2*a)

D[X] = (pi-3)/a^2

```python
import numpy as np
from numpy.random import random
from numpy import arcsin
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
def gen_rand(a,size):
    v = [arcsin(random())/a for t in range(size)] 
    return v
```

## 1. Придумать функцию, удовлетворяющую свойствам функции распределения случайной величины (указать их). Найти оценки параметров этого распределения Сгенерировать выборки и найти оценки этих параметров.

```python
a = 10 
t = gen_rand(a,100000)
print('Calc M: ',np.mean(t),'Expected M: ',(np.pi-2)/(2*a))
print('Calc D: ',np.var(t),'Expected D: ',(np.pi-3)/(a**2))
sns.distplot(t)
```

## 2. Изучить закон распределения Коши (закон арктангенса). Убедиться экспериментально, что математическое ожидание этой случайно величины не существует.
