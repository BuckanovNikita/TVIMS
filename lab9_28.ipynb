# Исследование случайных векторов и их характеристик.

```python
import numpy as np
from numpy import pi as pi
import seaborn as sns
from numpy.random import multivariate_normal
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Сгенерировать выборки 2-х мерных случайных величин. Выделить случай независимых, зависимых, коррелированных компонент. Дать графическую иллюстрацию.

### Независимые случайные величины

```python
mean = [0,0]
cov = [[1000,0],[0,1000]]
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

### Линейно зависимые компоненты cor(X,Y) = 1

```python
mean = [0,0]
cov = [[1,1],[1,1]]
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

### Коррелирующие компоненты

```python
mean = [0,0]
cov = [[[5,t],[t,1]] for t in np.linspace(0,np.sqrt(5),10)]
for q in cov:
    plt.figure()
    print(q[0][1])
    x, y = np.random.multivariate_normal(mean, q , 5000).T
    plt.plot(x, y, 'x')
    plt.axis('equal')
    plt.show()
```

## 2. Экспериментально показать, что зависимые компоненты могут оказаться некоррелируемыми.

Рассмотрим величину равномерно рапределённую на [-pi;pi]


```python
N = 10000
x = [np.cos(2*pi*phi-pi) for phi in np.random.random(N)]
y = [np.sin(2*pi*phi-pi) for phi in np.random.random(N)]
np.corrcoef(x,y)
```
