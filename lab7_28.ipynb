# Центральная предельная теорема и следствия из нее.

```python
import numpy as np
import numpy.random as rand
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Изучить выборки из сумм независимых одинаково распределенных случайных величин, удовлетворящих условиям ЦПТ. Сделать выводы. Проиллюстрировать результатами.

```python
size = [10,100,1000,10000,100000]
binomial = [rand.binomial(10000,0.45,t) for t in size]
for t in binomial:
    plt.figure()
    sns.distplot(t)
```

```python
chisquare = [rand.chisquare(10,t) for t in size]
for t in chisquare:
    plt.figure()
    sns.distplot(t)
```

```python
stdexp = [rand.standard_exponential(t) for t in size]
for t in stdexp:
    plt.figure()
    sns.distplot(t)
```

```python
poisson = [rand.poisson(10,t) for t in size]
for t in poisson:
    plt.figure()
    sns.distplot(t)
```
